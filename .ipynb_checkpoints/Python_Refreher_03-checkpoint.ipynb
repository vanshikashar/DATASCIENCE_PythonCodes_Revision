```python
# File handling in Python
# Opening a file
#Reading files 
#Using read()- Read entire file

file= open("file1.txt", "r")
content= file.read()
print(content)
file.close()  
```

```python
#Using readline()- Read Line by Line

file= open("file1.txt", "r")
line1= file.readline()
print(line1)
file.close()
```

```python
#using readlines()- read all lines as list
file= open("file1.txt", "r")
lines= file.readlines()
print(lines)
file.close()
```

```python
# Writing to files
#using write()- overwrites existing content

file= open("file1.txt", "w")
file.write("Hello from India")
file.close()
```

```python
# using writelines()- write multiple lines

lines= ["Hello\n", "Welcome to India\n", "Let's Learn Python"]
file= open("file1.txt", "w")
file.writelines(lines)
file.close()
```

```python
#Appending to file

file= open("file1.txt", "a")
file.write("\nThis is additional line to show appending in file")
file.close()
```

```python
# using with statement 

with open("file1.txt", "r") as file:
    content= file.read()
    print(content)
```

```python
# checking if file exists

import os

if os.path.exists("file1.txt"):
    print("File exists!")
else:
    print("File not found!")
```

```python
#deleting a file

import os

if os.path.exists("file1.txt"):
    os.remove("file1.txt")
    print("File deleted")
else:
    print("File does not exist")
```
