```python
# Tuples and tuple methods
#Creating a tuple

# Empty tuple
empty_tuple = ()
# Tuple with elements
numbers = (1, 2, 3, 4, 5)
# Mixed data types
mixed_tuple = (1, "Hello", 3.14, True)
# Single element tuple (comma is necessary)
single_element = (42,)

print(empty_tuple)
print(numbers)
print(mixed_tuple)
print(single_element)
```

```python
# Accessing tuple elements

my_tuple = (10, 20, 30, 40)
# Indexing
print(my_tuple[1]) 
# Slicing
print(my_tuple[1:3])
```

```python
# Tuple packing and unpacking

# Packing
person = ("Vanshika", 20, "Engineer")
# Unpacking
name, age, profession = person
print(name) 
print(age) 
print(profession)
```

```python
# Set and Set Methods
#Creating a set

# Empty set (must use set(), not {})
empty_set = set()
print(empty_set)
# Set with elements
numbers = {1, 2, 3, 4, 5}
print(numbers)
# Mixed data types
mixed_set = {1, "Hello", 3.14, True}
print(mixed_set)
# Creating a set from a list
unique_numbers = set([1, 2, 2, 3, 4, 4, 5])
print(unique_numbers)
```

```python
# Example

set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Union – combines all unique elements
print(set1 | set2)
print(set1.union(set2))

# Intersection – common elements
print(set1 & set2)
print(set1.intersection(set2))

# Difference – in set1 but not in set2
print(set1 - set2)
print(set1.difference(set2)) 

# Symmetric Difference – in either set, but not both
print(set1 ^ set2)
print(set1.symmetric_difference(set2))
```

```python
# Dictionary and dictionary methods
#Creating a dictionary

# Empty dictionary
empty_dict = {}
print(empty_dict)

# Dictionary with key-value pairs
student = {
"name": "Alice",
"age": 25,
"grade": "A"
}
print(student)

# Using dict() constructor
person = dict(name="John", age=30, city="New York")
print(person)
```

```python
# Accessing dictionary elements

# Using keys
print(student["name"]) 

# Using get() (avoids KeyError if key doesn't exist)
print(student.get("age")) 

print(student.get("height", "Not Found")) 
```

```python
# Example

student = {"name": "Vani", "age": 25, "grade": "A+"}

# Adding a new key-value pair
student["city"] = "London"

# Updating an existing value
student["age"] = 21

# Removing an item
student.pop("grade")

# Iterating over a dictionary
for key, value in student.items():
    print(key, ":", value)
```
